```python
import pyswarms as ps
import matplotlib.pyplot as plt
from pyswarms.utils.functions import single_obj as fx
from pyswarms.utils.plotters import plot_cost_history
```

zadanie 1

```python
# Set-up hyperparameters
options = {'c1': 0.5, 'c2': 0.3, 'w':0.92}

# Call instance of GlobalBestPSO
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2,
                                    options=options)

cost_history = optimizer.cost_history

# Perform optimization
stats = optimizer.optimize(fx.sphere, iters=100)
best_cost = stats[0]

# na ktorej iteracji osignelismy minimum ?
number_of_iteration = 0

for i in cost_history:
    if i != best_cost:
        number_of_iteration += 1
# # Plot!
plot_cost_history(cost_history)
plt.show()
print(number_of_iteration)
print(cost_history)
```

udało się osiągnąć: "best cost: 2.5942093178452526e-05, best pos: [0.00489788 0.00139745]"
w ktorej iteracji? 88

```python
from pyswarms.utils.plotters import plot_surface
from pyswarms.utils.plotters.formatters import Mesher
from pyswarms.utils.plotters.formatters import Designer
# Run optimizer
options = {'c1':0.5, 'c2':0.3, 'w':0.9}
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options)
# historia kosztów i pozycji

### trzeba zrobić tak:
optimizer.optimize(fx.sphere, iters=50)

pos_history_3d = m.compute_history_3d(optimizer.pos_history)
d = Designer(limits=[(-1,1), (-1,1), (-0.1,1)], label=['x-axis', 'y-axis', 'z-axis'])
#tworzenie animacji
m = Mesher(func=fx.sphere)
animation3d = plot_surface(pos_history=pos_history_3d, # Use the cost_history we computed
                           mesher=m, designer=d,       # Customizations
                           mark=(0,0,0))               # Mark minima
animation3d.save('plot2.gif', writer='imagemagick', fps=10)
```
